```python
import matplotlib.pyplot as plt
```

```python
names = ["James","Mac","Sam","Tom","Nick"]
sal = [45000,54000,67000,23000,34000]

data = {
    "names" : names,
    "sal" : sal
}
```

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(10,4))
plt.bar(data["names"], data["sal"], color='green')
plt.title("Salary of Employees")
plt.xlabel("Names")
plt.ylabel("Salary")
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(10,4))
plt.barh(data["names"], data["sal"], color='green')
plt.title("Salary of Employees")
plt.ylabel("Names")
plt.xlabel("Salary")
plt.show()
```
